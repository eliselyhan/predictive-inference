```python
from google.colab import drive
drive.mount('/content/drive')

import sys
sys.path.append('/content/drive/MyDrive/Colab Notebooks/NeurIPS2024')

import numpy as np
import pandas as pd
import pickle

from realdata import *

import os


# load data
path = '/content/drive/MyDrive/Colab Notebooks/NeurIPS2024'
df_raw = pd.read_csv(path + '/Transactions.csv')
encoding = 'label'
df = preprocess(df = df_raw, encoding = encoding)
```

```python
#algo = 'RF'
algo = 'XGB'


D = 10
M = int(np.ceil(826 / D))
list_t = np.array(range(1, M)) * D

prop = [0.3, 0.1]
window_sizes_train = [1, 16, 256, 1024]
window_sizes_infer = [1, 16, 256, 1024]
weights = [0.99, 0.9, 0.5, 0.25]
alpha = 0.1
delta = 0.1

test_next = False




idx_seeds = np.arange(71, 101)
le = idx_seeds[0]
results = []

for (i, s) in enumerate(idx_seeds):
    seed = s * 100
    data = splitting(df0 = df, prop = prop, seed = seed)
    results_seed = test_online(data, algo, list_t, window_sizes_train, window_sizes_infer, weights, test_next, alpha, delta, seed)
    results.append(results_seed)
    print('seed index {} finished'.format(s))

    if s % 10 == 0:
        filename = path + '/{}_{}_{}_{}.pkl'.format(algo, le, s, test_next)
        with open(filename, 'wb') as f:
            pickle.dump(results, f)
        print('intermediate results saved: ', le, s)
        results = []
        le = s + 1


```
